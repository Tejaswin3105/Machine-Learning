import dependencies

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
movies_data=pd.read_csv('movies1.csv')
movies_data.head(2)
```

```python
movies_data.shape
```

```python
movies_data.columns
```

```python
selected_columns=['genres','keywords','title','cast','director']
selected_columns
```

replacing null values with null strings

```python
for i in selected_columns:
    movies_data[i]=movies_data[i].fillna(' ')
```

combining all the selected columns in one column

```python
comined_features=movies_data['genres']+movies_data['keywords']+movies_data['title']+movies_data['cast']+movies_data['director']
```

```python
comined_features
```

```python
comined_features[0]
```

converting the text data into the feature vectors

```python
vectorizer=TfidfVectorizer()
```

```python
feature_vectors=vectorizer.fit_transform(comined_features)
```

```python
feature_vectors
```

```python
print(feature_vectors)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(feature_vectors)
```

```python
similarity
```

getting a movie name from the user

```python
movie_name=input("Enter your favourite movie name:")
```

```python
list_of_all_titles=movies_data['title'].tolist()
```

```python
list_of_all_titles
```

```python
find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)
```

```python
find_close_match
```

```python
close_match=find_close_match[0]
```

```python
close_match
```

find the index of the movie with title

```python
index_of_the_movie=movies_data[movies_data.title==close_match]['index'].values[0]
```

```python
index_of_the_movie
```

getting list of similar movies

```python
similarity_score=list(enumerate(similarity[index_of_the_movie]))
```

```python
similarity_score
```

sorting the movies based on their similarity

```python
sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)
```

```python
sorted_similar_movies
```

print the name of similar movies based on their index

```python
print("Movies suggested by the given user:\n")

i=1
for movie in sorted_similar_movies:
    index=movie[0]
    title_from_index=movies_data[movies_data.index==index]['title'].values[0]
    if (i<20):
        print(i,'.',title_from_index)
        i+=1
```
